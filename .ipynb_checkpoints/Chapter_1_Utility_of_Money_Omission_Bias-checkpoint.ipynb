### 1.1. Utility and Risk Aversion

Traditional economic model of consumer choice

- seek to describe consumers' behaviours when making financial decisions

- assume individuals are rational and they make decisions to maximise their expected utility 



Implications of the graph:

- More money is undoubtedly better than less: U(10) > U(5)
- The incremental (marginal) value of an additional dollar gets smaller as our wealth increases: U(5) – U(0) > U(10) – U(5)
- The probability-weighted sum of the utilities for each outcome is often less than the utility of the expected value of the outcomes


**Application: St Peterbug Paradox**

Imagine you are going to play a game with two rules:
1. An unbiased coin is tossed until it lands on Tails
2. The player is paid \\$2 if Tails comes up on the opening toss, \\$4 if Tails first appears on the second toss, \\$8 if Tails first appears on the third toss, etc.

The expected value is 

$$
\text{Expected Value} = \sum^{\infty}_{i=1} \frac{1}{2^i} \cdot 2^i = \infty
$$

However, **should you pay infinite amount of money to pay this game?**

No, because the amount one should pay should be measured by the utility of the game, not the same as the expected value.

$$
\text{Utility} = \sum^{\infty}_{i=1} \frac{1}{2^i} \cdot U(2^i) \ll \infty
$$

Suppose one's utility function is $U(X) = X^{0.7}$, the amount he would pay for this game would be:

```python
import numpy as np
np.sum([0.5**(i) * ((2**i)**0.7) for i in range(1020)])
```

### 1.3. Axioms (Assumptions) about the human behaviours

A number of axioms (assumptions about behavior that traditional
economists view as self-evident) that are implied by this model of consumers’ indifference curves and budget constraints.


#### Axiom 1: Dominance, or "More is Better"

> All else equal, more of a particular good is better than less.

Violations of this axiom are common, and many relate to a set of biases known collectively as the **“Disposition Effect.”**


#### Axiom 2: Invariance, or "Consistency"

> A decision-maker should not be affected by the way alternatives are presented. That is, her preference between option A and option B should not change based on the language used to describe the two.

This axiom sounds universally applicable, but we will see that it is one of the most violated axioms. We are in fact strongly influenced by how options are presented to us. 

We will see numerous examples of violation of this axiom throughout this course. These examples fit within the set of biases generally referred to as **Framing” issues** or **“Preference Reversals.”**


#### Axiom 3: Indepence, or "Cancellation"

> Choices should be independent of proportional reductions.
That is, introducing a third option that has no bearing on the choice between the first two should not affect your earlier choice.

As a more intuitive example, consider the following:

> A man comes into a delicatessen and asks what kinds of sandwiches they have. The attendant answers that they have roast beef and chicken. The patron thinks for a minute and then selects the roast beef. The server then says, “Oh, I forgot to mention, we also have tuna.” The patron responds, “Well, in that case I guess I’ll
have chicken.”

Violations of this axiom are often related to a bias known as the **“Certainty Effect.”**


## 2. Omission Bias

### Key Features


- Why anticipation of future regret, and a preference not to act but to let nature take its course, can drive us to make irrational decisions

-  Why our risk-taking tendencies can vary considerably, depending on whether our investment portfolio is out-performing or under-performing

- Why Prospect Theory, a utility model proposed by psychologists Kahnemann & Tversky in the 1970s, reflects human behavior far more accurately than the so-called “Rational” model that we looked at earlier.


### 2.1. Regret Theory and Omission Bias

**Regret Theory**: People make some decisions in response to the extent of our **anticipated regret** if the decision subsequently goes against them.


**Anticipated regret may be intensified by Omission Bias**: we would prefer not to be the active agent of the undesirable consequence.

### 2.2. Prospect Theory

**Prospect Theory** gives a utility function as a more realistic model of how we respond to gains and losses:

- Risk aversion over gains
- Risk seeking over losses
- Loss aversion



**Concave (risk averse) over gains:** Further incremental gains add even smaller amounts of extra “value”

**Convex (risk seeking) over losses:** The incremental pain of additional dollar
losses gets smaller the more we lose.

**Loss Aversion:** in the difference in the slopes of the positive and negative curves; We feel greater pain from a loss than we feel pleasure for an equal-sized gain
